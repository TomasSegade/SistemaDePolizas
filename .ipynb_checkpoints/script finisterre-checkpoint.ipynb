```python
import os
import requests
import pandas as pd
import PyPDF2
import win32print
import win32api
import time
import shutil
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
```

```python

# Función para imprimir páginas específicas de un archivo PDF en una impresora específica
def imprimir_pdf(pdf_file, impresora, pages):
    for page_num in pages:
        win32api.ShellExecute(0, "print", pdf_file, f'/d:"{impresora}" /p{page_num}', ".", 0)
        time.sleep(2)  # Agrega una pausa entre impresiones para evitar conflictos de impresión
        

ruta_descargas = 'C:/Users/Tomas/Desktop/script_python/polizas_finisterre'#MODIFICAR EN PC ///  hacer que se cree ne el programa
ruta_controlador = 'C:/Users/Tomas/Desktop/script_python/chromedriver-win64/chromedriver.exe'#MODIFICAR EN PC /// esta deberia instalarse en la misma carpeta que el programa
ruta_temporal = 'C:/Users/Tomas/Desktop/script_python/temp'  # Carpeta temporal para almacenar los archivos descargados


# Configurar el servicio del controlador de Chrome
service = Service(ruta_controlador)

# Opciones de navegación
opciones = webdriver.ChromeOptions()
prefs = {'download.default_directory': ruta_descargas}
opciones.add_experimental_option('prefs', prefs)
opciones.add_argument('--start-maximized')
opciones.add_argument('--disable-extensions')

# Crear el controlador de Chrome con las opciones y el servicio
driver = webdriver.Chrome(service=service, options=opciones)

# Leer datos desde el archivo CSV
datos_csv = pd.read_csv('C:/Users/Tomas/Desktop/script_python/archivoCSV_FINISTERRE.csv')#ESTA RUTA DEBERIA QUEDAR PREDEFINIDA CUANDO SE CREA QUE LA BUSQWUE AHI

# Abrir la página web de FINISTERRE
driver.get('https://web01.finisterreseguros.com/cot01/cotizador.html')

# Esperar a que el elemento ACCEDER sea clickeable y luego hacer clic
elemento_automotores = WebDriverWait(driver, 30).until(
    EC.element_to_be_clickable((By.XPATH, "//button[contains(@onclick, \"document.getElementById('contact').style.display='block'\")]"))
)
elemento_automotores.click()

# PRODUCTOR
campo_acceso = WebDriverWait(driver, 10).until(
    EC.visibility_of_element_located((By.ID, "username"))
)
campo_acceso.send_keys("prod419")

# PASSWORD
campo_acceso = WebDriverWait(driver, 10).until(
    EC.visibility_of_element_located((By.ID, "password"))
)
campo_acceso.send_keys("prod419")

# Esperar a que el elemento ENVIAR sea clickeable y luego hacer clic
elemento_enviar = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, "//button[contains(@onclick, \"init('NULL');move();\")]"))
)
elemento_enviar.click()

# Esperar a que el elemento PRODUCCION sea clickeable y luego hacer clic
enlace_produccion = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, "//span[@class='menuItemText' and text()=' Producción']/parent::a"))
)
enlace_produccion.click()

# Esperar a que el elemento DOMINIO sea clickeable y luego hacer clic
enlace_consulta_dominios = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, "//a[@class='menuItem' and contains(@onclick, 'CONNECT METHOD LCPWA005 P')]"))
)
enlace_consulta_dominios.click()

# Iterar sobre las filas del DataFrame
for indice, fila in datos_csv.iterrows():
    # Obtener el dominio de la fila actual
    dominio_actual = fila['Dominios']

    # Esperar a que el campo sea visible y luego ingresarlo
    campo_dominio = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.ID, "pol"))
    )

    # Limpiar el campo antes de ingresar el nuevo valor
    campo_dominio.clear()

    # Ingresar el dominio actual
    campo_dominio.send_keys(str(dominio_actual))

    # Esperar a que el botón sea clickeable y luego hacer clic
    boton_aceptar = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'ACEPTAR')]"))
    )

    # Esperar a que el elemento que está interceptando el botón desaparezca
    WebDriverWait(driver, 10).until_not(
        EC.visibility_of_element_located((By.ID, "loader"))
    )
    boton_aceptar.click()

    # Esperar a que el último elemento de la tabla sea clickeable
    try:

        # Esperar a que la tabla se cargue completamente
        WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.XPATH, "//table[@id='example']"))
        )
        elemento_ultimo = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//table[@id='example']/tbody/tr[last()]"))
        )

        # Hacer clic en el ultima poliza
        elemento_ultimo.click()

        # Esperar a que el elemento sea visible
        nombre_asegurado_element = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.XPATH, "//td[@class='td30b' and preceding-sibling::td[text()='Asegurado']]"))
        )

        # Extraer nombre del asegurado
        nombre_asegurado = nombre_asegurado_element.text.split('-')[1].strip()

        # Esperar a que el elemento sea visible
        propuesta_interna = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.XPATH, "//td[@class='td30b' and preceding-sibling::td[text()=' Propuesta Interna']]"))
        )

        # Extraer el valor de la Propuesta Interna
        propuesta_interna = propuesta_interna.text.strip()

        # Esperar a que el botón de impresión sea clickeable y luego hacer clic
        boton_impresion = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(@onclick, \"init('print');\")]"))
        )
        boton_impresion.click()

        # Esperar a que el enlace de la póliza sea clickeable y luego hacer clic
        enlace_poliza = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//a[@class='menuItem' and contains(@href, \"init('P')\")]"))
        )
        enlace_poliza.click()

        # Esperar un breve período de tiempo para que aparezca el popbox
        try:
            popbox = WebDriverWait(driver, 3).until(EC.visibility_of_element_located((By.CLASS_NAME, "popbox_container")))

            # Si el popbox está presente, hacer clic en el botón X para cerrarlo
            boton_cerrar_popbox = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//button[contains(@onclick, \"popbox.close('LCMSG')\")]"))
            )
            boton_cerrar_popbox.click()
        except TimeoutException:
            # Si el popbox no aparece, no hacemos nada
            pass

        # Esperar un breve período de tiempo para que aparezcan las ventanas emergentes
        try:
            # Esperar hasta que haya al menos dos ventanas abiertas (la ventana original y las ventanas emergentes)
            WebDriverWait(driver, 10).until(lambda driver: len(driver.window_handles) >= 2)

            # Obtener el identificador de la ventana original
            ventana_original = driver.window_handles[0]

            # Iterar sobre todas las ventanas emergentes, excepto la ventana original, y cerrarlas
            for ventana in driver.window_handles[1:]:
                driver.switch_to.window(ventana)
                driver.close()

            # Cambiar de nuevo a la ventana original
            driver.switch_to.window(ventana_original)
        except:
            # Si no aparecen ventanas emergentes, no hacemos nada
            pass

        # Construir el enlace con el número de póliza dinámico
        url_archivo = f"https://web01.finisterreseguros.com/pdfdocs/POL{propuesta_interna}P.PDF"
        
        # Descargar el archivo
        nombre_archivo = f"{ruta_descargas}/{nombre_asegurado}.pdf"
        
        # Verificar si el archivo original ya existe en la carpeta de descargas
        if os.path.exists(nombre_archivo):
            # Si ya existe, buscar un nombre único agregando un número al final
            indice = 1
            while True:
                # Construir un nuevo nombre de archivo
                nuevo_nombre = f"{ruta_descargas}/{nombre_asegurado}_{indice}.pdf"
                # Verificar si el nuevo nombre ya está en uso
                if not os.path.exists(nuevo_nombre):
                    # Si el nombre no está en uso, utilizarlo y salir del bucle
                    nombre_archivo = nuevo_nombre
                    break
                # Incrementar el índice para probar el siguiente nombre
                indice += 1

        with open(nombre_archivo, "wb") as archivo:
            contenido_archivo = requests.get(url_archivo).content
            archivo.write(contenido_archivo)

        # Verificar el tamaño del archivo descargado (en bytes)
        tamanio_archivo_bytes = os.path.getsize(nombre_archivo)

        # Definir el límite de tamaño (200KB en bytes)
        limite_tamanio_bytes = 200000

        # Verificar si el tamaño del archivo es menor al límite
        if tamanio_archivo_bytes < limite_tamanio_bytes:
            # Eliminar el archivo descargado si es demasiado pequeño
            os.remove(nombre_archivo)
            print(f"La póliza de {nombre_asegurado} no está refacturada.")
        else:
            print(f"Póliza de {nombre_asegurado} descargado exitosamente.")
            
            # Leer el archivo PDF descargado
            with open(nombre_archivo, "rb") as pdf_file:
                # Leer el archivo 
                reader = PyPDF2.PdfReader(pdf_file)
                                
                # Configurar la impresora
                impresora = win32print.GetDefaultPrinter()
                
                # Crear una copia del archivo con la primera página y renombrarla
                nombre_archivo_frente = os.path.join(ruta_descargas, f"{nombre_asegurado}_FRENTE.pdf")
                primera_pagina = reader.pages[0]
                writer = PyPDF2.PdfWriter()
                writer.add_page(primera_pagina)
                with open(nombre_archivo_frente, "wb") as frente_file:
                        writer.write(frente_file)
                    
                # Buscar la hoja que contiene la frase "detalle de cuotas" y crear una copia renombrada
                for page_num, page in enumerate(reader.pages, start=1):
                    text = page.extract_text()
                    if "detalle de cuotas" in text.lower():
                        nombre_archivo_factura = os.path.join(ruta_descargas, f"{nombre_asegurado}_FACTURA.pdf")
                        writer = PyPDF2.PdfWriter()
                        writer.add_page(page)
                        with open(nombre_archivo_factura, "wb") as factura_file:
                            writer.write(factura_file)
                        break
                
                # Imprimir el archivo creado con la primera página
                imprimir_pdf(nombre_archivo_frente, impresora, [1])
                time.sleep(1)  # Pausa antes de eliminar el archivo

                # Imprimir el archivo creado con la frase encontrada
                imprimir_pdf(nombre_archivo_factura, impresora, [1])
                time.sleep(1)  # Pausa antes de eliminar el archivo

                # Eliminar los archivos creados
                os.remove(nombre_archivo_frente)
                os.remove(nombre_archivo_factura)
                
        # SALIMOS DE LA POLIZA PARA BUSCAR OTRA
        boton_salida = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[@title='Salida']"))
        )

        # Hacer clic en el botón de salida
        boton_salida.click()

        # Esperar a que el botón sea clickeable
        boton_x = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(@onclick, \"popbox.close('LTVPAT01')\")]"))
        )

        # Hacer clic en el botón 'x'
        boton_x.click()

        # Esperar a que el campo sea visible y luego ingresar un nuevo dominio
        campo_dominio = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.ID, "pol"))
        )

        # Limpiar el campo antes de ingresar el nuevo valor
        campo_dominio.clear()

    except TimeoutException:
        try:
            # Si no se encuentra ningún elemento en la tabla, hacer clic en el botón "x"
            boton_x = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//button[contains(@onclick, \"popbox.close('LTVPAT01')\")]"))
            )
            boton_x.click()

            # Esperar a que aparezca el bloque mencionado
            bloque_importante = driver.find_element(By.XPATH, "//div[@class='w3-container w3-white']")

            # Si aparece el bloque, hacer clic en el botón correspondiente
            boton_bloque = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//button[contains(@onclick, \"popbox.close('LCPWA040')\")]"))
            )
            boton_bloque.click()

            # Esperar a que el elemento DOMINIO sea clickeable y luego hacer clic
            enlace_consulta_dominios = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//a[@class='menuItem' and contains(@onclick, 'CONNECT METHOD LCPWA005 P')]"))
            )
            enlace_consulta_dominios.click()

            continue  # Continuar con la siguiente línea del CSV

        except TimeoutException:
            # Si no aparece el bloque, continuar con la siguiente línea del CSV
            print("No se encontraron resultados para el dominio:", dominio_actual)
            continue

# Cerrar el navegador al finalizar
driver.close()
```
